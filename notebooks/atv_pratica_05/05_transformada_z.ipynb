## 1. Transformada Z simbólica (lcapy)

A biblioteca `lcapy` permite calcular a transformada Z de forma simbólica, de modo análogo ao `ztrans()` do MATLAB.

### Exemplo 1: Transformada Z da rampa

Calcule a transformada Z de $x[n] = nT_s$ (sinal rampa amostrado).

```python
import lcapy as lc
from lcapy.discretetime import n

Ts = lc.symbol('T_s', positive=True)
xn = n * Ts
print('x[n] =', xn)

Xz = xn.ZT()
print('X(z) =', Xz)
```

### Exemplo 2: Transformadas de sinais clássicos

Verifique os pares da tabela de transformadas Z para os seguintes sinais.

```python
from lcapy import UnitStep, UnitImpulse
from lcapy.discretetime import n

a = lc.symbol('a')
w0 = lc.symbol('omega_0')

sinais = {
    'delta[n]': UnitImpulse(n),
    'u[n]': UnitStep(n),
    'a^n u[n]': a**n * UnitStep(n),
    'n*a^n u[n]': n * a**n * UnitStep(n),
    'cos(w0*n) u[n]': lc.cos(w0*n) * UnitStep(n),
    'sin(w0*n) u[n]': lc.sin(w0*n) * UnitStep(n),
}

for nome, xn in sinais.items():
    print(f'Z{{ {nome} }} = {xn.ZT()}')
    print()
```

## 2. Transformada Z inversa

O método `IZT()` calcula a transformada Z inversa, equivalente ao `iztrans()` do MATLAB.

### Exemplo 3: Transformada inversa por frações parciais

Obtenha a transformada Z inversa de:

$$X(z) = \frac{z + 1}{z^2 + 0{,}3z + 0{,}02}$$

```python
from lcapy.discretetime import z

Xz = (z + 1) / (z**2 + 0.3*z + 0.02)
print('X(z) =', Xz)

xn = Xz.IZT()
print('x[n] =', xn)
```

**OBS:** O termo `UnitImpulse(n)` corresponde ao impulso unitário discreto $\delta[n]$, definido como $\delta[n] = 1$ se $n = 0$ e $\delta[n] = 0$ se $n \neq 0$.

## 3. Função de transferência discreta (python-control)

A biblioteca `python-control` permite criar e manipular modelos de sistemas LIT, de forma análoga ao `tf()` e `zpk()` do MATLAB.

### Exemplo 4: Criando G(z) com tf()

```python
import control

num_Gz = [1, 2, 7]
den_Gz = [1, -0.3, 0.02]
Ts = 0.5

# Sem definir o período de amostragem
G = control.tf(num_Gz, den_Gz, True)
print(G)

# Com o período de amostragem definido
G = control.tf(num_Gz, den_Gz, Ts)
print(G)
```

### Exemplo 5: Polos, zeros e ganho com zpk()

```python
K = 10
zer_Gz = [-0.1, -0.2]
pol_Gz = [-0.3, -0.4]
Ts = 0.5

G = control.zpk(zer_Gz, pol_Gz, K, Ts)
print(G)
print('Polos:', control.poles(G))
print('Zeros:', control.zeros(G))
```
